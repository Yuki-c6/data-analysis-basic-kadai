# 0.学生成績管理システム

  

あなたは学校の成績管理システムを作成することになりました。以下の要件を満たすPythonプログラムを作成してください。

  

1. 以下の学生と成績（3つのテスト結果）を持つ辞書 `grades` を作成してください：

   - 田中: [80, 75, 90]

   - 佐藤: [65, 70, 85]

   - 鈴木: [90, 95, 88]

   - 高橋: [72, 68, 81]

  

2. 成績の平均点を計算する関数 `calculate_average(scores)` を作成してください。

  

3. 合格した学生（平均点が特定の点数以上）をリストアップする関数 `get_passed_students(grades, passing_score)` を作成してください。

  

4. 新しい学生と成績を追加する関数 `add_new_student(grades, student, scores)` を作成してください。

  

5. 全生徒の成績と平均点を表示する関数 `print_all_grades(grades)` を作成してください。

  

6. メイン処理として、以下の操作を行ってください：

   a. 現在の全生徒の成績と平均点を表示する

   b. 合格点を75点とし、合格した学生をリストアップして表示する

   c. 新しい学生 "山田" と成績 [78, 82, 88] を追加する

   d. 更新後の全生徒の成績と平均点を表示する

  

注意：

- 関数を適切に使用してプログラムを構造化してください。

- コードにはコメントを付けて、各部分の目的を説明してください。

- 平均点は小数点第1位まで表示してください。

````python
# ライブラリのインポート
import pandas as pd
import numpy as np

# データの作成
grades: dict[str, list[int]] = {'田中': [80, 75, 90], '佐藤': [65, 70, 85], '鈴木': [90, 95, 88], '高橋': [72, 68, 81]}


# 平均を計算する定義
def calculate_average(scores: list):
    x = np.mean(scores)
    return  round(x, 1)


# 合格した学生をリストアップする関数を定義
def get_passed_students(grades: dict, passing_score: int):

    # 出力用のリストを定義
    passing_student: str = []

    # 辞書の中をループ
    for key, value in grades.items():

        # scoreを平均値に変換する
        mean_score = calculate_average(value)
        grades[key] = mean_score

        # passing_score以上の場合のみリストに追加
        if mean_score >= passing_score:
            passing_student.append(key)

    return passing_student


# 新しい学生と成績を追加する関数を定義
def add_new_student(grades: dict, student: str, scores:list[int]):
    grades[student] = scores
    return grades


# 全生徒の成績と平均値を表示する関数を定義
def print_all_grades(grades: dict):
    for key , value in grades.items():
        mean_score = calculate_average(value)
        grades[key] = mean_score
        print(f'{key} : {mean_score}')


# 現在の状況を表示する
print('現在')
print_all_grades(grades)

# 合格者を表示する
print('\n合格者')
print(get_passed_students(grades, 75))

# 新しい学生を追加する
add_new_student(grades, '山田', [78, 82, 88])

# 更新後の状況を表示する
print('\n更新後')
print_all_grades(grades)
````

````python
# データ作成
import numpy as np

x = np.linspace(0, 10, 100)

y1 = np.sin(x)

y2 = np.cos(x)

y3 = np.tan(x)
````

````python
import matplotlib.pyplot as plt

# 画像サイズの変更
fig, ax = plt.subplots(figsize=(10, 10))

# グラフの書式を変更
ax.plot(x, y1, color='red', lw = 1, marker='.', linestyle='-', label='y1')
ax.plot(x, y2, color='blue', lw = 2, marker=',', linestyle='--', label='y2')
ax.plot(x, y3, color='green', lw =3, marker='x', linestyle=':', label='y3')
ax.legend()

# 凡例の表示
plt.show()
````
