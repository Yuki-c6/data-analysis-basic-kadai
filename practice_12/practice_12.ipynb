## 実践12-1 (pandasの復習)
* 「7. 予測モデルの評価」で実行したclassification_reportをdataframeに格納してください。行名・列名もわかるようにしてください。その後csvとして保存してください。
* 実務でもdataframeに格納してcsvへ保存するケースはあります。

````python
# ライブラリのインポート
import pandas as pd

# データの取得
from sklearn.datasets import load_breast_cancer
dataset = load_breast_cancer()

# データ内容の確認
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
df['class'] = dataset.target
display(df.head())
df.shape
df.describe()
````

````python
# データ欠損の確認
display(df.info())
df.hist(figsize=(20, 15), bins=30)
````

````python
# 学習データとテストデータへの分割
X = df.drop(columns=['class']).to_numpy()
y = df['class'].to_numpy()

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
````

````python
# 予測モデルの学習
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(random_state=0)
model.fit(X_train, y_train)
````

````python
# モデルの評価
y_pred = model.predict(X_test)

from sklearn.metrics import classification_report
cf_report = classification_report(y_test, y_pred, output_dict=True)

# classification_reportのDataFrame変換
cf_df = pd.DataFrame(cf_report)
display(cf_df)

# csvで保存
cf_df.to_csv('classification_report.csv')
````

## 実践12-2(matplotlib, seabornの復習 & 一部発展)
* load_breast_cancerを講義テキストでは
```
df.hist(figsize=(20, 15), bins=30)
```
として実施しました。
* これをseabornのstripplotを使って同様に表示してください。その際、全特徴量を1つのfigとして表示し、図を保存してください。
* ポイント: subplotsを用いること & for文を使ってstripplotをおこなう。
* 参考) 実務でもこのように全特徴量に対して一斉の可視化を行うケース(pandasの可視化だけでは対応できないケース)はあります。一度ロジックを組み立てておけば汎用的に使えるので覚えておくと便利です。
* 更に発展させると、この処理のベースを関数/クラスのメソッド化しておき、dataframeを渡すだけにしておくと更に利便性が高まります。


````python
# ライブラリのインポート
import pandas as pd

# データの取得
from sklearn.datasets import load_breast_cancer
dataset = load_breast_cancer()
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
df['class'] = dataset.target
display(df.head())
print(df.shape)
display(df.describe())

# データの欠損の確認
df.info()
````

````python
# seabornでstripplotを行う
import seaborn as sns
import matplotlib.pyplot as plt
import math

# カラム数を算出
total_column = len(df.columns)
plot_num = math.ceil(math.sqrt(total_column))

# サブプロットを定義
fig, axes = plt.subplots(plot_num, plot_num, figsize=(20,15), tight_layout=True)
axes_flat = axes.ravel()

# 不要な範囲を削除
remove_start = total_column % plot_num
if remove_start != 0:
    for ax_num in range(remove_start, plot_num):
        fig.delaxes(axes[plot_num - 1, ax_num])

# stripplotを設定し、表示
for col, ax in zip(df.columns, axes_flat):
    sns.stripplot(x=col, data=df, ax=ax)
    ax.set
plt.show()
````

## 実践12-3
- 「7. 予測モデルの評価」の流れで混同行列を作成し表示してください。(classification_reportの元となる行列で実務でも利用します)
- 混同行列をdataframe(index, column名付与)して表示・保存してください。

````python
# ライブラリのインポート
import pandas as pd

# データの取得
from sklearn.datasets import load_breast_cancer
dataset = load_breast_cancer()

# データ内容の確認
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
df['class'] = dataset.target
display(df.head())
display(df.shape)
display(df.info())
display(df.describe())
````

````python
# 学習データとテストデータの分割
X = df.drop(columns=['class']).to_numpy()
y = df['class'].to_numpy()

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
````

````python
# 予測モデルの学習
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(random_state=0)
model.fit(X_train, y_train)
````

````python
# 予測モデルの評価
y_pred = model.predict(X_test)

# 混同行列の作成
from sklearn.metrics import confusion_matrix
labels = [1, 0]
cm = confusion_matrix(y_test, y_pred, labels=labels)

# データフレームの作成
index_labels = [["test", "test"], [True, False]]
columns_labels = [["pred", "pred"], [True,False]]
cm_df = pd.DataFrame(cm, columns=columns_labels, index=index_labels)

# 混同行列の表示と保存
display(cm_df)
cm_df.to_csv("confusion_matrix.csv")
````

## 実践12-4(マルチクラス分類)
* irisデータを取得してください。
```
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data
y = iris.target
```
* irisデータに対して学習・推論を実施してください。モデルは任意で構いませんが、木構造以外のアルゴリズムの場合は標準化を視野に入れてください(厳密にはデータによっては標準化が不要な場合もありますが、基本は実施すべきです)。木構造の場合はexport_text等で条件分岐を出力してください。
* 混同行列や評価指標を算出してください。


````python
import pandas as pd

# データの取得
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data
y = iris.target

# データ内容の確認
df = pd.DataFrame(X,columns=iris.feature_names)
df['target'] = y
display(df.head())
display(df.info())
display(df.shape)
display(df.describe())
````

````python
# 分布の確認
df.hist(figsize=(10, 8), bins=30)
````

````python
# 学習データの分割
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
````

````python
# モデルの学習と予測
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(random_state=0)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# 混同行列の表示
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm_df = pd.DataFrame(cm, index=[["test", "test", "test"], [0, 1, 2]], columns=[["predict", "predict", "predict"], [0, 1, 2]])
print("混同行列")
display(cm_df)

# モデルの評価
from sklearn.metrics import classification_report
report = classification_report(y_test, y_pred, output_dict=True)
cf_df = pd.DataFrame(report)
print("\nモデルの評価")
display(cf_df)
````

````python
# 条件分岐構造を表示
from sklearn.tree import export_text
names = iris.feature_names

print(export_text(model, decimals=3, feature_names=names))
````
