## 実践8-1
* irisデータ（非常に有名な花の種類(アヤメ)のデータセット）を用いて基礎集計をしてください。
* 基礎集計からこのデータの説明をしてください。
* 基礎集計後、train, testに分割しtestデータを推論してください。
* データの読み込みは以下です。
```
from sklearn.datasets import load_iris
iris = load_iris()
X, y = iris.data, iris.target
```
* モデルはDecisionTreeClassifierと課題と同じランダムフォレストの２種類を使ってください。
* accuracyを算出してください。

````python
from sklearn.datasets import load_iris
iris = load_iris()
X, y = iris.data, iris.target
````

````python
print(iris.DESCR)
````

````python
import pandas as pd
df = pd.DataFrame(data=X, columns=iris.feature_names)
df.head()
````

````python
df.info()
````

````python
"""
このデータは全部で150行あり、インデックスをの除く4列で構成されいる。
それぞれの列において空白値はなく、float64型で格納されている。
1行目：がくの長さ(cm)　平均5.843333　標準偏差0.828066　最大値7.900000　最小値4.300000
2行目：がくの幅(cm)　　平均3.057333　標準偏差0.435866　最大値4.400000　最小値2.000000
3行目：花びらの長さ(cm)　平均3.758000　標準偏差1.765298　最大値6.900000　最小値1.000000
4行目：花びらの幅(cm)　　平均1.199333　標準偏差0.762238　最大値2.500000　最小値0.100000
"""
display(df.describe())
df.shape
````

````python
# 訓練データとテストデータに分割
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=5)
print(X.shape, X_train.shape, X_test.shape, y.shape, y_train.shape, y_test.shape)
````

````python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

# 各モデルのインスタンス化
model_decision_tree = DecisionTreeClassifier(random_state=3)
model_random_forest = RandomForestClassifier(random_state=3)

# 各モデルを学習
model_decision_tree.fit(X_train, y_train)
model_random_forest.fit(X_train, y_train)
````

````python
from sklearn.metrics import accuracy_score

# 予測値を代入
y_pred_decision_tree = model_decision_tree.predict(X_test)
y_pred_random_forest = model_random_forest.predict(X_test)

# 予測値を確認
print(f"決定木\n{y_pred_decision_tree}")
print(f"ランダムフォレスト\n{y_pred_random_forest}")

# 正解率を算出して表示
print(f"決定木の正解率:{accuracy_score(y_test, y_pred_decision_tree)}")
print(f"ランダムフォレストの正解率:{accuracy_score(y_test, y_pred_random_forest)}")
````

## 実践8-3
* がんデータを読み込んでください。
```
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```
* iris同様、決定木とランダムフォレストのモデルを用いて推論してください。
* 正解率(accuracy)を算出してください
* (応用) 適合率(precision), 再現率(recall), F1値を算出してください。（ヒント: "講義資料の予測モデルの評価"にI/F記載あり）


````python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
X, y = cancer.data, cancer.target
````

````python
print(cancer.DESCR)
````

````python
import pandas as pd

# データの確認
df = pd.DataFrame(data=X, columns=cancer.feature_names)
display(df.head())
````

````python
df.info()
````

````python
display(df.describe())
print(df.shape)
````

````python
from sklearn.model_selection import train_test_split

# 学習データとテストデータに分割
X_test, X_train, y_test, y_train = train_test_split(X, y, train_size=0.3, random_state=5)
print(X.shape, X_train.shape, X_test.shape, y.shape, y_train.shape, y_test.shape)
````

````python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

# モデルをインスタンス化
model_decision_tree = DecisionTreeClassifier(random_state=3)
model_random_forest = RandomForestClassifier(random_state=3)

# モデルを学習
model_decision_tree.fit(X_train, y_train)
model_random_forest.fit(X_train, y_train)
````

````python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# 予測値を代入
y_pred_decision_tree = model_decision_tree.predict(X_test)
y_pred_random_forest = model_random_forest.predict(X_test)

# 予測値を確認
# print(f"決定木\n{y_pred_decision_tree}")
# print(f"ランダムフォレスト\n{y_pred_random_forest}")

# 正解率を表示
print("正解率")
print(f"決定木:{accuracy_score(y_test, y_pred_decision_tree)}")
print(f"ランダムフォレスト:{accuracy_score(y_test, y_pred_random_forest)}")

# 適合率を表示
print("\n適合率")
print(f"決定木:{precision_score(y_test, y_pred_decision_tree)}")
print(f"ランダムフォレスト:{precision_score(y_test, y_pred_random_forest)}")

# 再現率を表示
print("\n再現率")
print(f"決定木:{recall_score(y_test, y_pred_decision_tree)}")
print(f"ランダムフォレスト:{recall_score(y_test, y_pred_random_forest)}")

# F値を表示
print("\nF値")
print(f"決定木:{f1_score(y_test, y_pred_decision_tree)}")
print(f"ランダムフォレスト:{f1_score(y_test, y_pred_random_forest)}")
````
