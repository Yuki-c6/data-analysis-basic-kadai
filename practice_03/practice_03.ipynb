##演習3-1
* 行列の積(アダマール積)を計算して値を返す関数を定義し、複数の値を呼び出し結果を確認してください。

```python
import numpy as np

# アダマール積（要素同士の積を計算し返す関数を定義)
def calculate_hadamard_product(array_1, array_2):
    return array_1 * array_2


# 計算し、出力を行う
array_A = np.array([[1, 2, 3], [4, 5, 6]])
array_B = np.array([[1, 2, 3], [4, 5, 6]])
print(calculate_hadamard_product(array_A, array_B))

array_C = np.array([[7, 8, 9], [10, 11, 12]])
array_D = np.array([[6, 6, 6], [3, 3, 3]])
print(calculate_hadamard_product(array_C, array_D))

array_E = np.array([[1, 3, 5], [7, 9, 13]])
array_F = np.array([[13, 9, 7], [5, 3, 1]])
print(calculate_hadamard_product(array_E, array_F))

```

##演習3-2
* 行列を転置(ヒント: .T)するメソッドを持つクラスを作成してください。コンストラクタにはその行列を渡し、クラス内で保持して転置メソッドで利用してください。


```python
import numpy as np

# クラスを定義
class Matrix:
    # コンストラクタを定義
    def __init__(self,matrix):
        self.matrix = matrix

    # 転置メソッドを定義
    def transposition(self):
        return self.matrix.T

# インスタンス化を行う
matrix = np.array([[1, 2, 3], [4, 5, 6]])
input =Matrix(matrix)

# 転置前と転置後をそれぞれ出力する
print(input.matrix)
print(input.transposition())

# print(matrix.transpose())
# print(np.transpose(matrix))
```

## 演習3-6
* 2次元配列（5×5）を1次元配列へ変換してください。
* その際、2次元配列はfor文×2を用いて作成してください。

```python
import numpy as np

# 格納用のリストを定義
row = []
matrix = []

# リスト内に数値を格納
for i in range(5):
    for j in range(5):
         row.append(j)
    matrix.append(row)
    row = []

# ndarray形式に変換
array = np.array(matrix)

# 1次元配列に変換し出力
flat_array = np.ravel(array)
print(flat_array)
```
