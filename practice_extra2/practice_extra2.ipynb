## 1. strftime
次のタイムスタンプを "YYYY年MM月DD日 HH時MM分" 形式で表示してください。


```
timestamp = pd.Timestamp('2023-05-15 14:30:00')
```

````python
import pandas as pd
timestamp = pd.Timestamp('2023-05-15 14:30:00')
````

````python
timestamp = timestamp.strftime('%Y年%m月%d日 %H時%M分')
print(timestamp)
````

## 2. strptime
以下の文字列形式の日付をdatetimeオブジェクトに変換してください。変換にはdatetime.strptimeメソッドを使用してください。



```
date_string = "2023-04-01 14:30"
```

変換後、以下の情報を表示してください：

* 年
* 月
* 日
* 時
* 分

ヒント：

from datetime import datetimeを使用してdatetimeモジュールをインポートします。
strptimeメソッドの第二引数にフォーマット文字列を指定します。

````python
date_string = '2023-04-01 14:30'
````

````python
from datetime import datetime
date_time = datetime.strptime(date_string, '%Y-%m-%d %H:%M')
````

````python
print(f'年 : {date_time.year}')
print(f'月 : {date_time.month}')
print(f'日 : {date_time.day}')
print(f'時 : {date_time.hour}')
print(f'分 : {date_time.minute}')
````

2. 集計操作をおこない商品と地域ごとの売上の平均、最大値、最小値を計算してください。


````python
import pandas as pd
import numpy as np

# サンプルデータの作成
np.random.seed(42)
data = {
    '日付': pd.date_range(start='2023-01-01', end='2023-12-31', freq='D'),
    '商品': np.random.choice(['A', 'B', 'C'], size=365),
    '地域': np.random.choice(['東京', '大阪', '名古屋'], size=365),
    '売上': np.random.randint(1000, 10000, size=365)
}

df = pd.DataFrame(data)

# データの確認
print(df.head())
````

````python
mean_pivot = pd.pivot_table(df, values='売上', index='商品', columns='地域')
mean_pivot = mean_pivot.reset_index()
mean_pivot.columns.name = None
print('平均')
display(mean_pivot)
````

````python
max_pivot = pd.pivot_table(df, values='売上', index='商品', columns='地域', aggfunc='max')
max_pivot = max_pivot.reset_index()
max_pivot.columns.name = None
print('最大')
display(max_pivot)
````

````python
min_pivot = pd.pivot_table(df, values='売上', index='商品', columns='地域', aggfunc='min')
min_pivot = min_pivot.reset_index()
min_pivot.columns.name = None
print('最小')
display(min_pivot)
````
